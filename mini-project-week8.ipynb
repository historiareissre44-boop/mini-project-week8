```python
"""
Day 40 Activity: Matrix Operations Exercise
Tasks:
1) Build a simple linear transformation Y = X @ W + b
2) Avoid loops; use vectorized operations
3) Print shapes and verify output
"""

import numpy as np

# Provided inputs (edit if you want different values)
X = np.array([
    [1.0, 0.5],
    [2.0, -1.0],
    [0.0, 3.0],
])
W = np.array([
    [0.2, -0.1, 0.5],
    [0.7, 0.3, -0.2],
])
b = np.array([0.1, 0.0, -0.3])

# TODO: Compute Y = X @ W + b
Y = X @ W + b
# TODO: Print shapes
print("Shape of X:", X.shape)
print("Shape of W:", W.shape)
print("Shape of b:", b.shape)
print("Shape of Y:", Y.shape)

print("\nOutput Y:")
print(Y)
```
